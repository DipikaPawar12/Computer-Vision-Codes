### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

### Create Classifiers

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Open Image

```python
img = cv2.imread('Images/eye_face.jpg')
```

### Fix Image

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
faces =  face_classifier.detectMultiScale(fix_img, 1.3 ,5)
```

### If No Face Detected

```python
if faces is ():
    print('No face is detected')
```

### It's a Kind of Magic

```python
def detect_face_eye(fix_img):
    
    face_rects = face_classifier.detectMultiScale(fix_img, 1.3, 5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                    (x,y),
                    (x+w,y+h),
                    (255, 0,0),
                    7)
        eyes_rects = eye_classifier.detectMultiScale(fix_img, 1.3, 5)
        for(ex, ey, ew, eh) in eyes_rects:
            cv2.rectangle(fix_img,
                            (ex,ey),
                          (ex+ew, ey+eh),
                         (0,0,255),
                         5)
            
    return fix_img
```

```python
result = detect_face_eye(fix_img)
plt.imshow(result)
```
