### Import Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
```

### Show Image

```python
plt.imshow(image)
```

### Fix Image

```python
fix_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Show Image

```python
plt.imshow(fix_image)
```

### Create Classifier

```python
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### It's a Kind of Magic

```python
def detect_eyes(fix_image):
    eyes_rectangle = eye_classifier.detectMultiScale(fix_image)
    
    for (x, y, w, h) in eyes_rectangle:
        cv2.rectangle(fix_image,
                     (x,y),
                     (x+w, y+h),
                     (255, 255, 255),
                     10)
    return fix_image
```

### Show Results

```python
result = detect_eyes(fix_image)
plt.imshow(result)
```
