### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifier

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
fix_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

### Detect Faces

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

### No Faces Detected

```python
if faces is ():
    print('No faces found')
```

### It's a kind of Magic

```python
def detect_face(fix_image):
    face_rects = face_classifier.detectMultiScale(fix_image)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(fix_image,
                     (x,y),
                     (x+w, y+h),
                     (255, 0, 0),
                     10)
    return fix_image

```

```python
result =detect_face(fix_image)
plt.imshow(result)
```
